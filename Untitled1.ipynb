```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score, LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix

# Configuration for plots
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# ==========================================
# 1. Data Loading and Preparation
# ==========================================
# Load Iris Dataset
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
df_iris = pd.DataFrame(X_iris, columns=iris.feature_names)
df_iris['species'] = [iris.target_names[i] for i in y_iris]

# Load Breast Cancer Dataset
data_cancer = load_breast_cancer()
X_cancer, y_cancer = data_cancer.data, data_cancer.target
df_cancer = pd.DataFrame(X_cancer, columns=data_cancer.feature_names)

# Display dataset structure
print("--- Iris Dataset Head ---")
print(df_iris.head())
print("\n--- Breast Cancer Dataset Info ---")
df_cancer.info()

# ==========================================
# 2. K-NN Experiment: Confusion Matrix (Iris)
# ==========================================
# Split data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

# Train K-NN model with k=3
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Visualize Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Confusion Matrix - Iris (k=3)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# ==========================================
# 3. K-NN Experiment: Impact of k (Breast Cancer)
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(X_cancer, y_cancer, test_size=0.3, random_state=42)

# Standardize features (Mean=0, Std=1) - Essential for distance-based algorithms
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# Evaluate accuracy for k ranging from 1 to 20
acc_list = []
k_values = range(1, 21)
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_s, y_train)
    acc_list.append(knn.score(X_test_s, y_test))

# Plot accuracy vs. k
plt.figure()
plt.plot(k_values, acc_list, marker='o')
plt.title('Effect of k on Accuracy (Breast Cancer)')
plt.xlabel('Value of k')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.show()

# ==========================================
# 4. Normalization Impact Analysis
# ==========================================
# K-NN on Raw Data (k=5)
knn_raw = KNeighborsClassifier(n_neighbors=5)
knn_raw.fit(X_train, y_train)
acc_raw = knn_raw.score(X_test, y_test)

# K-NN on Normalized Data (k=5)
knn_s = KNeighborsClassifier(n_neighbors=5)
knn_s.fit(X_train_s, y_train)
acc_s = knn_s.score(X_test_s, y_test)

# Compare results
plt.figure(figsize=(6, 5))
plt.bar(['Raw Data', 'Normalized Data'], [acc_raw, acc_s], color=['#ff9999','#66b3ff'])
plt.ylim(0.85, 1.0)
plt.title('Accuracy Comparison: Raw vs Normalized (k=5)')
plt.text(0, acc_raw, f"{acc_raw:.4f}", ha='center', va='bottom')
plt.text(1, acc_s, f"{acc_s:.4f}", ha='center', va='bottom')
plt.show()

# ==========================================
# 5. Dimensionality Reduction (PCA)
# ==========================================
# Apply PCA to project data into 2D subspace
pca = PCA(n_components=2)
# Fit PCA on standardized dataset
X_pca = pca.fit_transform(scaler.fit_transform(X_cancer))

plt.figure()
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_cancer, cmap='viridis', alpha=0.7, edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection (Breast Cancer Subspace)')
plt.legend(handles=scatter.legend_elements()[0], labels=list(data_cancer.target_names))
plt.show()

# ==========================================
# 6. Model Validation (Cross-Validation & LOOCV)
# ==========================================
# 10-Fold Cross-Validation on Breast Cancer dataset
scores = cross_val_score(KNeighborsClassifier(n_neighbors=5), scaler.fit_transform(X_cancer), y_cancer, cv=10)
print(f"\n10-Fold CV Mean Accuracy: {np.mean(scores):.4f}")

# Leave-One-Out Cross-Validation (LOOCV) on Iris dataset
loo = LeaveOneOut()
loo_scores = cross_val_score(KNeighborsClassifier(n_neighbors=3), X_iris, y_iris, cv=loo)
error_rate = 1 - np.mean(loo_scores)
print(f"LOOCV Error Rate (Iris Dataset): {error_rate:.4f}")

# ==========================================
# 7. Data Distribution Visualization
# ==========================================
sns.pairplot(df_iris, hue="species")
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()

# ==========================================
# 8. SVM Classifier Experiment
# ==========================================
results = []
kernels = ['linear', 'rbf', 'poly']

print("\n--- SVM Kernels Performance Comparison ---")
for k in kernels:
    svc = SVC(kernel=k)
    svc.fit(X_train_s, y_train) # Train on normalized data
    acc = svc.score(X_test_s, y_test)
    results.append({'Kernel': k, 'Accuracy': acc})

print(pd.DataFrame(results))
```
